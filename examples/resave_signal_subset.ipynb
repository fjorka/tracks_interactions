```python
import os
import zarr
from ome_zarr.io import parse_url
from ome_zarr.reader import Reader
from ome_zarr.writer import write_image
from numcodecs import Blosc
```

```python
source_path = r'P:\Kasia\tracking\TrackGardener\E6_C0.zarr'
destination_path = r'D:\tracks_interactions\test_data\E6_C0_level_1.zarr'
```

```python
reader = Reader(parse_url(source_path))
image_node = list(reader())[0].data[1]
image_node
```

```python
store = parse_url(destination_path,mode='w').store
root = zarr.group(store=store)
compressor = Blosc(cname='lz4', clevel=5, shuffle=Blosc.SHUFFLE)
write_image(image=image_node, group=root, axes="tyx", storage_options=dict(chunks=(10,512,512),compressor=compressor))
```
